## ***Ejercicio Fechas Publicación***

Teniendo en cuenta que existe una Fecha de Publicación del Proceso y una fecha de última publicación, realice la limpieza de formato de dichas variables. Defina una nueva variable en la que se calcule la diferencia de éstas dos fechas y responda si incluiría las dos fechas dentro de la fuente de datos.

*Utilice estadísticas descriptivas para responder la pregunta*

```python
for col in df_SECOP.columns:
    num_missing = df_SECOP[col].isnull().sum()
    pct_missing = np.mean(df_SECOP[col].isnull())
    #print(col," - ",round(pct_missing*100, 2),"% - ",num_missing)
    print('{} - {}% - {}'.format(col,round(pct_missing*100, 2),num_missing))
```

```python
df_SECOP['Fecha de Publicacion del Proceso'] = pd.to_datetime(df_SECOP['Fecha de Ultima Publicación'], format='%m/%d/%Y')
df_SECOP['Fecha de Ultima Publicación'] = pd.to_datetime(df_SECOP['Fecha de Ultima Publicación'], format='%m/%d/%Y')
```

```python
dif_fechas = df_SECOP['Fecha de Ultima Publicación'] - df_SECOP['Fecha de Publicacion del Proceso']
```

```python
dif_fechas.describe()
```

```python
dif_fechas1 = df_SECOP['Fecha de Publicacion del Proceso'] - df_SECOP['Fecha de Ultima Publicación']
dif_fechas1.describe()
```

Realizando el análisis de la Fecha de publicación del proceso y una fecha de ultima de publicación,  y una vez realizado el proceso de limpieza se evidencia que no es necesario incluir las dos fechas dentro de la base de los datos.

## ***Ejercicio Tipo de Contrato y Subtipo de Contrato***
De acuerdo con las variables Contrato y Subcontrato, realice una tabla cruzada y responda ¿Vale la pena dejar ambas variables en la fuente de datos?

```python
punto2 = pd.crosstab(df_SECOP['Tipo de Contrato'],['Subtipo de Contrato'])
punto2
```

```python
punto2b = pd.crosstab(df_SECOP['Subtipo de Contrato'],['Subtipo de Contrato'])
punto2b
```

```python
l=df_SECOP.groupby(by=['Tipo de Contrato','Subtipo de Contrato']).sum().groupby(level=[0]).cumsum()
print(l)
```
